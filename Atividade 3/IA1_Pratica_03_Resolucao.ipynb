```python
import time
from random import randint
```

```python
# Definindo variaveis

agente = 'A'

limpo = 0

sujo = 1

ambiente = [[0,0,0],[0,0,0],[0,0,0]]
```

```python
def criarSalas(ambiente, qtd): #qtd = sala por row, sendo 3 rows perma
    qtd = int(qtd)
    rows = 3
    for i in range(qtd):
        for j in range (rows):
            valor = randint(0,1)
            print(str(valor))
            ambiente[i][j] = valor
    print("Posicionando agente.")
    ambiente[0][0] = agente
  
def posAgente(ambiente, agente): #encontra posição do agente
    for i in range(len(ambiente)): #percorre o ambiente
         for j in range(len(ambiente[i])): #percorre a row
            if(ambiente[i][j] == agente):
                return i,j
  
def checaSala(ambiente,row, sala):
    if ambiente[row][sala] == 0:
        return True #sala limpa
    else:
        return False #sala suja

def mudarPosicao(ambiente, rowAntiga, row, salaAntiga, sala, agente):
    ambiente[rowAntiga][salaAntiga] = 0
    ambiente[row][sala] = agente

def movimento(direcao, ambiente, agente):
    
    row, sala = posAgente(ambiente, agente)
    match direcao:
        case "esquerda": 
            if (sala == 0):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row, row, sala, sala-1, agente)
        case "direita": 
            if (sala == 2):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row, row, sala, sala+1, agente)
        case "cima":
            if (row == 0):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row-1, row, sala, sala, agente)
        case "baixo":
            if (row == 2):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row, row+1, sala, sala, agente)
        case _:
            print("Movimento inválido")
    rowNova, salaNova = posAgente(ambiente, agente)
    print("Nova posição: " + str((rowNova, salaNova)))
    # Checa direções para continuar movimento
    if salaNova - 1 >= 0 and not checaSala(ambiente, rowNova, salaNova - 1):
        movimento("esquerda", ambiente, agente)
    elif salaNova + 1 < 3 and not checaSala(ambiente, rowNova, salaNova + 1):
        movimento("direita", ambiente, agente)
    elif rowNova - 1 >= 0 and not checaSala(ambiente, rowNova - 1, salaNova):
        movimento("cima", ambiente, agente)
    elif rowNova + 1 < 3 and not checaSala(ambiente, rowNova + 1, salaNova):
        movimento("baixo", ambiente, agente)
    else:
        mostrar_na_tela(ambiente)
        return

def mostrar_na_tela(ambiente):
    
    for sala in ambiente:
        
        print(sala)
        
    
    print("\n")


criarSalas(ambiente, 3)
mostrar_na_tela(ambiente)  
movimento("direita", ambiente, agente)
```
