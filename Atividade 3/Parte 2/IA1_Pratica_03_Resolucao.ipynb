```python
def criarSalas(ambiente, qtd): #qtd = sala por row, sendo 3 rows perma
    qtd = int(qtd)
    rows = 3
    for i in range(qtd):
        for j in range (rows):
            valor = randint(0,1)
            ambiente[i][j] = valor
    print("Posicionando agente.")
    ambiente[0][0] = agente
  
def posAgente(ambiente, agente): #encontra posição do agente
    for i in range(len(ambiente)): #percorre o ambiente
        for j in range(len(ambiente[i])): #percorre a row
            if(ambiente[i][j] == agente):
                return i,j
  
def checaSala(ambiente,row, sala):
    if ambiente[row][sala] == 0:
        return True #sala limpa
    else:
        return False #sala suja


def mudarPosicao(ambiente, rowAntiga, row, salaAntiga, sala, agente):
    if (ambiente[row][sala] == 1):
        ambiente[row][sala] = 0
    ambiente[rowAntiga][salaAntiga] = 0
    ambiente[row][sala] = agente

def movimento(ambiente, agente):
   
    visitadas = []
    visitadas.append(posAgente(ambiente, agente))
   
    while (len(visitadas) < 9):
        row, sala = posAgente(ambiente, agente)  
        sujas = [] #prioridade para as salas sujas
        movs = [
            ("cima", (row-1, sala)),
            ("baixo", (row+1, sala)),
            ("esquerda", (row, sala-1)),
            ("direita", (row, sala+1))
        ]   
        for (mov, (rowMov,salaMov)) in movs:
            if (0 <= rowMov < 3 and 0 <= salaMov < 3) and (rowMov, salaMov) not in visitadas:
                if (ambiente[rowMov][salaMov] == 1):
                    suja = 0
                else:
                    suja = 1
                sujas.append((suja, mov, rowMov, salaMov))
        
        if not sujas:
            break
        
        sujas.sort()
        suja, mov, rowMov, salaMov = sujas[0]
        
        mudarPosicao(ambiente, row, rowMov, sala, salaMov, agente)
        visitadas.append((rowMov,salaMov))
            
        
    '''
    snake (eater) pattern 
    sinceramente, já que ele não volta, é melhor classificar como método Boustrophedon
    kept you waiting huh
    método boustrophedon sem priorização de salas sujas, mas é mais eficiente que o método anterior.
    deixando aqui pq foi resultado de pesquisa e gostei disso
    movs = ["baixo", "baixo", "direita", "cima", "cima", "direita", "baixo", "baixo"]
    for mov in movs:
        match mov:
            case "esquerda": 
                row, sala = posAgente(ambiente, agente)
                if (sala == 0):
                    print("Movimento: " + mov + " inválido")
                    continue  
                else:
                    mudarPosicao(ambiente, row, row, sala, sala-1, agente)
            case "direita": 
                row, sala = posAgente(ambiente, agente)
                if (sala == 2):
                    print("Movimento: " + mov + " inválido")
                    continue 
                else:
                    mudarPosicao(ambiente, row, row, sala, sala+1, agente)
            case "cima":
                row, sala = posAgente(ambiente, agente)
                if (row == 0):
                    print("Movimento: " + mov + " inválido")
                    continue 
                else:
                    mudarPosicao(ambiente, row, row-1, sala, sala, agente)
            case "baixo":
                row, sala = posAgente(ambiente, agente)
                if (row == 2):
                    print("Movimento: " + mov + " inválido")
                    continue 
                else:
                    mudarPosicao(ambiente, row, row+1, sala, sala, agente)
            case _:
                print("Movimento: " + mov + " inválido")
    mostrar_na_tela(ambiente)
    '''
    

def mostrar_na_tela(ambiente):
    
    for sala in ambiente:
        
        print(sala)
        
    
    print("\n")

def aspirador(ambiente, agente):
    criarSalas(ambiente, 3) #parametro qtd define salas por row, como já comentado.
    movimento(ambiente, agente)
    return ambiente
```
