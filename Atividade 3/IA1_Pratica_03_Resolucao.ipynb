```python
import time
from random import randint
```

```python
# Definindo variaveis

agente = 'A'

limpo = 0

sujo = 1

ambiente = [[0,0,0],[0,0,0],[0,0,0]]
```

```python
def criarSalas(ambiente, qtd): #qtd = sala por row, sendo 3 rows perma
    qtd = int(qtd)
    rows = 3
    for i in range(qtd):
        for j in range (rows):
            valor = randint(0,1)
            print(str(valor))
            ambiente[i][j] = valor
    print("Posicionando agente.")
    ambiente[0][0] = agente
  
def posAgente(ambiente, agente): #encontra posição do agente
    for i in range(len(ambiente)): #percorre o ambiente
         for j in range(len(ambiente[i])): #percorre a row
            if(ambiente[i][j] == agente):
                posicao = "Row " + str(i) + " Sala " + str(j)
                return i,j,posicao
            else:
                print("Procurando agente")  
  
def mudarPosicao(ambiente, row, sala, agente):
    ambiente[row][sala] = agente

def movimento(direcao, ambiente, agente):
    row, sala, pos = posAgente(ambiente, agente)
    match direcao:
        case "esquerda": 
            if (sala == 0):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row, sala-1, agente)     
        case "direita": 
            if (sala == 2):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row, sala+1, agente)
                     
        case "cima":
            if (row == 0):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row+1, sala, agente)
        case "baixo":
            if (row == 2):
                print("Movimento inválido")
            else:
                mudarPosicao(ambiente, row-1, sala, agente)
        case _:
            print("Movimento inválido")
    print("movimento concluido")
    
            
            
                     

def mostrar_na_tela(ambiente):
    
    for sala in ambiente:
        
        print(sala)
        
    
    print("\n")


criarSalas(ambiente, 3)
mostrar_na_tela(ambiente)  
movimento("direita", ambiente, agente)
posAgente(ambiente, agente)
```
